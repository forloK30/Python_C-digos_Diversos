### Código para calcular: Entropia Máxima (Hmax), Entropia (H), Redundância (R), Redundância Normalizada (r) e Quantidade de Informação (I(N))

- **Restrição:** A calculadora somente recebe lista de letras

- **Insights:** É possível em poucas linhas de código adaptar a calculadora para receber uma mensagem (**Dica:** use o FOR para percorrer a mensagem e adicione cada caractere na lista N caso não seja um espaço)

```python
import math

# Definição das funções
def Calc_Hmax(n):
    # Calcula a entropia máxima possível com base no número de elementos únicos
    return float("{:.4f}".format(math.log2(n))) #Hmax

def Calc_H(prob):
    # Calcula a entropia com base na probabilidade de cada elemento
    soma = 0
    for prob_elemento in prob:
        soma += float(prob[prob_elemento])*math.log2(float(prob[prob_elemento]))
    return float("{:.4f}".format(- soma))

def Calc_R(Hmax, H):
     # Calcula a redundância com base na entropia máxima e na entropia real
    return float("{:.4f}".format(Hmax - H))

def Calc_r(Hmax, H):
    # Calcula a redundância normalizada com base na entropia máxima e na entropia real
    return float("{:.4f}".format(100*(Hmax - H) / Hmax))

def Calc_I(H, k):
    # Calcula a quantidade de informação com base na entropia e na quantidade de elementos
    return float("{:.4f}".format(H*k))

# Recebe a sequência de elementos do usuário
elemento = str(input("Digite o elemento: "))

N = [] # lista de elementos
Rep = [] # lista de elementos únicos
prob = {} # dicionario para alocar as probabilidades

while(elemento != "fim"):
    print("Digite 'fim' para encerrar a sequência de elementos.\n")
    N.append(elemento.upper())
    elemento = str(input("Digite o elemento: "))
    
else:
    # Lista para testes (tirar # caso queira testar)
    #N = ['B', 'B', 'C', 'D', 'E', 'F', 'F', 'F', 'G', 'A', 'A', 'A', 'B', 'B', 'C', 'C', 'D', 'D', 'E', 'F', 'G', 'G', 'F', 'A', 'B', 'D', 'C', 'B', 'A']
    #N =  ['A', 'N', 'A', 'M', 'A', 'R', 'I', 'A', 'E', 'N', 'T', 'R', 'O', 'U', 'N', 'A', 'C', 'A', 'B', 'I', 'N', 'E']
    #N = ['O', 'U', 'V', 'I', 'R', 'A', 'M', 'D', 'O', 'I', 'P', 'I', 'R', 'A', 'N', 'G', 'A', 'A', 'S', 'M', 'A', 'R', 'G', 'E', 'N', 'S', 'P', 'L', 'A', 'C', 'I', 'D', 'A', 'S', 'D', 'E', 'U', 'M', 'P', 'O', 'V', 'O', 'H', 'E', 'R', 'O', 'I', 'C', 'O', 'O', 'B', 'R', 'A', 'D', 'O', 'R', 'E', 'T', 'U', 'M', 'B', 'A', 'N', 'T', 'E']
    
    print("\nA sequência de elementos informada foi (N): ", N)
    k = len(N) 
    print("A qtd de elementos é (k)", k)
    
    Rep = sorted(list(set(N)))
    print("\nA sequência de elementos únicos é (Rep): ", Rep)
    n = len(Rep)
    print("A qtd de elementos únicos é (n): ", n)
    
    print("\n")
    
    
    # Calcula a probabilidade de cada elemento único
    for elemento in Rep:
        print("Probabilidade: ", "p" + elemento.upper(), "=", N.count(elemento), "/", k)
        
        prob["p" + elemento.upper()] = float("{:.4f}".format(N.count(elemento) / k))
    print("\nAs probabilidades são: ", prob)
    
    
    # Calcula Hmax, H, R, r e I(N)
    Hmax = Calc_Hmax(n)
    print("\nA entropia máxima é (Hmax): ", Hmax)
    
    H = Calc_H(prob)
    print("A entropia é (H): ", H)
    
    R = Calc_R(Hmax, H)
    print("A redundância é (R): ", R)
    
    r = Calc_r(Hmax, H)
    print("A redundância normalizada é (r): ", r)
    
    I = Calc_I(H, k)
    print("A quantidade de informação é (I(N)): ", I)
```

## CALCULADORAS ALTERNATIVAS

#### CALCULADORA DE CADA CONTA DO SOMATÓRIO PARA O CÁLCULO DE ENTROPIA

```python
import math

continuar = "sim"
prob = 0

print("Digite o valor em decimal (0.0) ou fração (/)")
while continuar != "fim":
    prob = float("{:.4f}".format(eval(input("\nDigite a probabilidade para calcular seu log2: "))))
    log2 = float("{:.4f}".format(math.log2(prob)))

    n = float(input("Digite quantas vezes a probabilidade se repete: "))
    entropia_prob = n * prob * log2

    print("\nlog2 de", prob, "é:", log2)
    print("\n\nResultado:", entropia_prob)
    
    continuar = str(input("Digite 'fim' para encerrar: "))
```
